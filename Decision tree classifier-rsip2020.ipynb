```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.feature_selection import SelectFromModel
```

```python
dataset =pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset
```

```python
x= dataset.iloc[:,1:4].values
x
```

```python
y= dataset.iloc[:,4:].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
x[:,0]=lb.fit_transform(x[:,0])
```

```python
x
```

```python
feat_labels = ['Gender','Age','EstimatedSalary']
```

```python
from sklearn.model_selection import train_test_split                #previously cros_validation was used in sklearn
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from joblib import dump
dump(sc,"scalar.save")
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
dt.fit(x_train,y_train)
```

```python
import pickle
pickle.dump(dt,open('decision.pkl','wb'))
```

```python
for feature in zip(feat_labels, dt.feature_importances_):
    print(feature)
```

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dt, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
